```python
import numpy as np
import pandas as pd
import math
import pingouin as pg
```

```python
# Read in the BFI Acoustics data

bfia_df = pd.read_csv('BFI_Acoustics.csv', sep=',', header=0, index_col=0)
bfia_df
```

```python
# Read in the BFI Personality data

bfip_df = pd.read_csv('BFI_Personality.csv', sep=',', header=0, index_col=0)
bfip_df
```

```python
# Reverse Keying 

Personality_Arr = bfip_df.to_numpy()
Personality_Arr
```

```python
neg_cols = np.array([1,2,5,8,9,11,12,18,19])

for col in neg_cols:
    Personality_Arr[:,col] = 6 - Personality_Arr[:,col]

Personality_Arr
```

```python
Acoustics_Arr = bfia_df.to_numpy()
Acoustics_Arr
```

```python
# Score Matrix

Psych_score_mat = np.zeros((Personality_Arr.shape[0], 5))
for i in range(Personality_Arr.shape[0]):
    # Openness
    Psych_score_mat[i,0] = Personality_Arr[i,19] + Personality_Arr[i,14] + Personality_Arr[i,4] + Personality_Arr[i,9]
    
    # Conscientiousness
    Psych_score_mat[i,1] = Personality_Arr[i,7] + Personality_Arr[i,2] + Personality_Arr[i,12] + Personality_Arr[i,17]
    
    # Extraversion
    Psych_score_mat[i,2] = Personality_Arr[i,15] + Personality_Arr[i,0] + Personality_Arr[i,5] + Personality_Arr[i,10]
    
    # Agreeableness
    Psych_score_mat[i,3] = Personality_Arr[i,1] + Personality_Arr[i,11] + Personality_Arr[i,6] + Personality_Arr[i,16]
    
    # Neuroticism
    Psych_score_mat[i,4] = Personality_Arr[i,18] + Personality_Arr[i,8] + Personality_Arr[i,13] + Personality_Arr[i,3]
    
Psych_score_mat = Psych_score_mat / 4
Psych_score_mat
```

```python
Psych_df = pd.DataFrame(Psych_score_mat, columns=['Openness', 'Conscientiousness', 'Extraversion', 'Agreeableness', 'Neuroticism'])
Psych_df
Psych_df.to_csv('Psych_Scores.csv')
```
